## Azure Architecture Process

To ensure system is fast, secure, reliable and easy to maintain, it is required to follow these best practices:
1. Understand the system requirements
2. Understand the non-functional requirements
3. Map the components
4. Select the Technology stack.
5. Design the Architecture.
6. Write Architecture Document
7. Support Team

#### 1. Understand the system's requirements

- Understand the system goals
- Requirements = what the system should do?
- Like work flow, user interface elements, logical services etc...
- Usually defined by the system analyst who works with clients
- Might require multiple followup meetings to understand all the requirements

#### 2. Understand the non-functional requirements

- Define technical and service level attributes
- ie. # of users (concurrunt number of users), heavy Loads, Volumes of data, Performacne
- Clients might not know these details always.
- Much more important than regular requirements.
- it's architects job to formulate these requirements.
